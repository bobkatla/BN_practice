```python
import pandas as pd
import geopandas as gpd
```

```python
POA = gpd.read_file("../data/spatial/POA_2021_AUST_GDA2020_SHP.zip")
SA1_2021 = gpd.read_file("../data/spatial/SA1_2021_AUST_SHP_GDA2020.zip")
SA1_2011 = gpd.read_file("../data/spatial/vic_sa1_2011.gpkg")
SA1_2016 = gpd.read_file("../data/spatial/vic_sa1_2016.gpkg")
```

```python
SA1_2011_sim = SA1_2011[["SA1_MAINCODE_2011", "geometry"]]
SA1_2016_sim = SA1_2016[["SA1_MAINCODE_2016", "geometry"]]
SA1_2021_sim = SA1_2021[["SA1_CODE21", "geometry"]]
POA_2021_sim = POA[["POA_CODE21", "geometry"]]
```

```python
hh_seed_raw = pd.read_csv("..\data\source2\VISTA\SA\H_VISTA_1220_SA1.csv")
```

```python
hh_seed_sim = hh_seed_raw[["hhid", "homesa1"]]
hh_seed_sim["homesa1"] = hh_seed_sim["homesa1"].astype(str).str.replace(".0", "")
hh_seed_sim
```

```python
combine_spatial_11 = SA1_2011_sim.to_crs(POA_2021_sim.crs).sjoin(POA_2021_sim)
combine_spatial_11
```

```python
combine_spatial_16 = SA1_2016_sim.to_crs(POA_2021_sim.crs).sjoin(POA_2021_sim)
combine_spatial_16
```

```python
combine_spatial_21 = SA1_2021_sim.to_crs(POA_2021_sim.crs).sjoin(POA_2021_sim)
combine_spatial_21
```

```python
dict_map_SA1_POA_21 = dict(zip(combine_spatial_21["SA1_CODE21"], combine_spatial_21["POA_CODE21"]))
dict_map_SA1_POA_21
```

```python
dict_map_SA1_POA_11 = dict(zip(combine_spatial_11["SA1_MAINCODE_2011"], combine_spatial_11["POA_CODE21"]))
dict_map_SA1_POA_16 = dict(zip(combine_spatial_16["SA1_MAINCODE_2016"], combine_spatial_11["POA_CODE21"]))
dict_map_SA1_POA_21 = dict(zip(combine_spatial_21["SA1_CODE21"], combine_spatial_11["POA_CODE21"]))
```

```python
dict_fin_sa1_poa = {}
dict_fin_sa1_poa.update(dict_map_SA1_POA_11)
dict_fin_sa1_poa.update(dict_map_SA1_POA_16)
dict_fin_sa1_poa.update(dict_map_SA1_POA_21)
```

```python
# actual mapping
hh_seed_sim["POA"] = hh_seed_sim["homesa1"].apply(lambda x: dict_fin_sa1_poa[x])
hh_seed_sim
```

```python
hh_seed_sim.to_csv("hhid_sa1_poa.csv", index=False)
```
