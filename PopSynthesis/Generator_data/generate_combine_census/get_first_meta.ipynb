```python
import pandas as pd
import numpy as np
from collections import defaultdict

from PopSynthesis.Generator_data.generate_combine_census.utils import process_from_census_data
```

```python
geo_lev = "POA"
data_raw = process_from_census_data(geo_lev=geo_lev)
data_raw
```

```python
# Vehicle
```

```python
persons_cols = [x for x in data_raw.columns if "Person_" in x]
hh_cols = [x for x in data_raw.columns if "Dwelling_" in x]

pp_marg_raw = data_raw[persons_cols]
hh_marg_raw = data_raw[hh_cols]
```

```python
dict_check_hh = defaultdict(list)
for hcol in hh_marg_raw.columns:
    if geo_lev not in hcol:
        ori_att, state = hcol.split("__")
        type_df, att = ori_att.split("_")
        dict_check_hh[att].append(state)

data_hh_to_df = {
    "att_code_census": [],
    "att_description": [],
    "state_census": [],
    "att_sample": [],
    "state_sample": []
}
for att, states in dict_check_hh.items():
    att_code, *other = att.split(" ")
    att_des = ' '.join(other)
    for state in states:
        data_hh_to_df["att_code_census"].append(att_code)
        data_hh_to_df["att_description"].append(att_des)
        data_hh_to_df["state_census"].append(state)
        data_hh_to_df["att_sample"].append("")
        data_hh_to_df["state_sample"].append("")

df_meta_hh = pd.DataFrame(data_hh_to_df)
# df_meta_hh.to_csv("meta_data_hh_1.csv", index=False)
```

```python
dict_check_pp = defaultdict(list)
for pcol in pp_marg_raw.columns:
    if geo_lev not in pcol:
        ori_att, state = pcol.split("__")
        type_df, att = ori_att.split("_")
        dict_check_pp[att].append(state)

data_pp_to_df = {
    "att_code_census": [],
    "att_description": [],
    "state_census": [],
    "att_sample": [],
    "state_sample": []
}
for att, states in dict_check_pp.items():
    att_code, *other = att.split(" ")
    att_des = ' '.join(other)
    for state in states:
        data_pp_to_df["att_code_census"].append(att_code)
        data_pp_to_df["att_description"].append(att_des)
        data_pp_to_df["state_census"].append(state)
        data_pp_to_df["att_sample"].append("")
        data_pp_to_df["state_sample"].append("")

df_meta_pp = pd.DataFrame(data_pp_to_df)
# df_meta_pp.to_csv("meta_data_pp_1.csv", index=False)
```
