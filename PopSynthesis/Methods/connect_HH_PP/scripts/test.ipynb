```python
import pandas as pd
```

```python
census_sa1 = pd.read_csv("../data/census_SA1.csv")
census_sa1.columns
```

```python
census_sa1["Total_dwelings"].sum()
```

```python
cols = [x for x in census_sa1.columns if "Num_MVs_per_dweling" in x]
cols.remove("Num_MVs_per_dweling_Tot")
cols
```

```python
def f(row):
    val = row["Total_dwelings"] - row[cols].sum()
    return val if val >= 0 else 0
census_sa1["new_Num_MVs_NS"] = census_sa1.apply(f, axis=1)
```

```python
census_sa1["Total_dwelings"].sum() - (census_sa1["Num_MVs_NS"].sum() + census_sa1[cols].sum().sum())
```

```python
census_sa1["Total_dwelings"].sum() - (census_sa1["new_Num_MVs_NS"].sum() + census_sa1[cols].sum().sum())
```
